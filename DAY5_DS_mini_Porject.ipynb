# **Mini Project: E-commerce Sales Analysis**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Superstore.csv', encoding='latin-1')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
## check missing values

df.isnull().sum()
```

```python
# Top 10 product by Sales

top_product = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
top_product = pd.DataFrame(top_product)
top_product.plot(kind='bar', figsize=(10,6))
plt.title('Top 10 Product by Sales')
plt.xlabel('Product Name')
plt.ylabel('Sales')
plt.show()
```

```python
# Sales trend over months

df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Month'] = df['Order Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Sales'].sum()

monthly_sales.plot(kind='line', marker='o', figsize=(12,6))
plt.title('Monthly Sales Trend')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```
